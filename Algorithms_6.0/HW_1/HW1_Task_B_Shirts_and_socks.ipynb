## Описание задачи

Как известно, осенью и зимой светает поздно, и так хочется утром ещё хоть немного поспать, а не идти в школу! Некоторые школьники готовы даже одеваться, не открывая глаз, лишь бы отложить момент пробуждения. Вот и Саша решил, что майку и носки он вполне может вытащить из шкафа на ощупь с закрытыми глазами и только потом включить свет и одеться.

В шкафу у Саши есть два ящика. В одном из них лежит \(A\) синих и \(B\) красных маек, а в другом — \(C\) синих и \(D\) красных пар носков. Саша хочет, чтобы и майка, и носки были одного цвета. Он вслепую вытаскивает \(M\) маек и \(N\) пар носков. Задача состоит в том, чтобы найти минимальное суммарное количество предметов одежды \((M + N)\), которые Саша должен вытащить, чтобы среди них гарантированно оказались майка и носки одного цвета.

**Задача**: Определите минимальное количество маек \(M\) и пар носков \(N\), которые Саша должен взять, чтобы среди них гарантированно оказалась майка и носки одного цвета.

### Входные данные

Программа получает на вход четыре целых неотрицательных числа \(A\), \(B\), \(C\), \(D\), записанных в отдельных строках:
- \(A\) — количество синих маек,
- \(B\) — количество красных маек,
- \(C\) — количество синих носков,
- \(D\) — количество красных носков.

Все числа не превосходят \(10^9\). Гарантируется, что в шкафу есть одноцветный комплект из майки и носков.

### Выходные данные

Программа должна вывести два числа: количество маек \(M\) и количество пар носков \(N\), которые должен взять Саша. Необходимо, чтобы среди \(M\) маек и \(N\) пар носков обязательно нашлась одноцветная пара, при этом сумма \(M + N\) должна быть минимальной.



```python
def solution (a,b,c,d):
    m, n = 10**9, 10**9

    # cases with 0
    if a == 0: return 1, c+1
    elif b == 0: return 1, d+1
    elif c == 0: return a+1, 1
    elif d == 0: return b+1, 1    

    # other cases
    if (a>b and c>d) or (a<b and c<d):
        m1, n1 = min(a,b)+1, min(c,d)+1
        if m1+n1 < m+n: m, n = m1, n1

    if max(a,b) <= max(c,d):
        m1, n1 = max(a,b)+1, 1
        if m1+n1 < m+n: m, n = m1, n1

    if max(a,b) > max(c,d):
        m1, n1 = 1, max(c,d)+1
        if m1+n1 < m+n: m, n = m1, n1

    return m, n
```

## Пояснение к решению

Для решения задачи необходимо рассмотреть различные случаи, чтобы минимизировать количество вытаскиваемых предметов:

1. **Проверка на наличие одного цвета**: Если в одном из ящиков нет маек или носков определённого цвета, Саша сразу может взять все предметы другого цвета и гарантированно получить комплект.

2. **Определение оптимального выбора**: Когда есть майки и носки обоих цветов, мы рассматриваем следующие варианты:
   - Если количество синих и красных маек совпадает по соотношению с количеством носков, то минимальное количество маек и носков можно определить, взяв минимальное количество предметов каждого цвета.
   - Если количество маек или носков одного цвета больше другого, то сначала берём максимум одного цвета и минимальное количество другого для гарантии одноцветного комплекта.
   - Для каждого случая вычисляется сумма \(M + N\), и выбирается тот вариант, который даёт минимальную сумму.

Эти условия позволяют Саше минимизировать количество вытаскиваемых предметов и гарантировать, что он получит комплект из маек и носков одного цвета.


## Пример

Теперь мы можем ввести свои данные для тестирования.

```python
# Ввод данных для теста
a = int(input("Введите A: "))
b = int(input("Введите B: "))
c = int(input("Введите C: "))
d = int(input("Введите D: "))

# Вызов функции и вывод результата
res = solution(a,b,c,d)
print("Ответ:", res[0],res[1])
```

### Примечания
В примере из условия в шкафу лежит \(A = 6\) синих маек и \(B = 2\) красных маек. Если взять 3 майки, то среди них обязательно найдётся синяя. В другом ящике лежит \(C = 7\) пар синих носков и \(D = 3\) пары красных носков. Если взять 4 пары, то среди них обязательно будет пара синих носков. Поэтому если взять вслепую 3 майки и 4 пары носков, то среди них обязательно найдётся одноцветный (синий) комплект из майки и носков.
